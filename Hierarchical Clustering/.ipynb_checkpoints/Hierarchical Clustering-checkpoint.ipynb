```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,3:].values
```

### Find the optimal number of clusters using Dendogram

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X, method = 'ward')) # minimum variance within the clusters.
plt.title('Dendrogram- Hierarchical Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, metric = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

```python
plt.scatter(X[y_hc == 0,0],X[y_hc == 0,1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1,0],X[y_hc == 1,1], s = 50, c = 'pink', label = 'Cluster 2')
plt.scatter(X[y_hc == 2,0],X[y_hc == 2,1], s = 50, c = 'magenta', label = 'Cluster 3')
plt.scatter(X[y_hc == 3,0],X[y_hc == 3,1], s = 50, label = 'Cluster 4')
plt.scatter(X[y_hc == 4,0],X[y_hc == 4,1], s = 50, c = 'cyan', label = 'Cluster 5')
plt.title("Agglomerative CLustering")
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
